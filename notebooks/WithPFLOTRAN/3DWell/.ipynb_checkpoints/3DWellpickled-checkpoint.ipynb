## Function that will run the model

**First we build each simulation case:** For each row in the Excel file, a PFLOTRAN input file is built.

```python
def buildSimulation(i=0):
  
    f = wholeTable.loc[i]
    caseName = f.loc[titleCase.tag] 
    
    ## Create a folder for the case
    current_folder = "./CASE" + "{0:02}".format(i+1)
    current_file = current_folder + "/" + caseName +".in"
    
    system("mkdir " + current_folder)
    
    ## Initialize PFLOTRAN model
    caseModel = mo.Model(
        templateFile=template_file,
        runFile = current_file
        )
       
    ## Copy template input file to folder
    caseModel.cloneTemplate()

    ## Replace tags for values in case
    for qp in listOfObjs:
        t = qp.tag
        if "dX" in t:
            value = mo.buildDXYZ(
                f.loc[Len.tag],
                f.loc[CellRatioX.tag],
                f.loc[nX.tag],
                hasBump=True)
        elif "dY" in t:
            value = mo.buildDXYZ(
                f.loc[Len.tag],
                f.loc[CellRatioY.tag],
                f.loc[nY.tag],
                hasBump=True)
        elif "dZ" in t:
            value = mo.buildDXYZ(
                f.loc[H.tag],
                f.loc[CellRatioZ.tag],
                f.loc[nZ.tag],
                hasBump=False)
        else:
            value = f.loc[t]

        qp.value = value
 
        caseModel.replaceTagInFile(qp)

    return caseModel
```

**Another function to build all the cases at once**

```python
def buildAllSimulations():
    caseModels = []
    for i in range(total_rows):
        caseModels.append(buildSimulation(i=i))
    return caseModels
```

```python
MassBalanceFiles = [groupFolder+"/"+f for f in listdir(groupFolder) if f[-4:] == ".dat"]
ProbeObservFiles = [groupFolder+"/"+f for f in listdir(groupFolder) if f[-4:] == ".tec"]
#print(MassBalanceFiles,ProbeObservFiles)
```

### Plotting stuff

```python
ConcentrationAtInlet = 1.00E-20 #mol/L
plt.figure(figsize=(12,4))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

for f in MassBalanceFiles:
    fdf = read_csv(f)
    t = fdf["Time [d]"]
    mV = -fdf["ExtractWell Vaq [mol/d]"]
    mW = -fdf['ExtractWell Water Mass [kg/d]']
    C = (mV/mW)/ConcentrationAtInlet
    ax1.plot(t,C,label=f[20:-8])
    
ax1.set_xlabel("Time [$d$]",fontsize="large")
ax1.set_ylabel("Rel Concentration   $\dfrac{C}{C_0}$",fontsize="large")
ax1.axhline(y=1.0,ls="dashed",c="teal",lw=1)
ax1.axhspan(ymin=-1.0E-7,ymax=1.0E-6,facecolor="pink",alpha=0.2)
ax1.set_yscale("log")
ax1.legend(title="From mass balance")

for f in ProbeObservFiles:
    fdf = read_csv(f)
    fieldName = lambda keys : [f for f in list(keys) if "Total Vaq" in f]
    t = fdf["Time [d]"]
    C = fdf[fieldName(fdf.keys())]/ConcentrationAtInlet
    ax2.plot(t,C,label=f[20:-8])
    
ax1.set_yscale("symlog",\
  linthresh=1.0E-6,subs=[1,2,3,4,5,6,7,8,9])
ax1.set_ylim([-1.0E-7,1.15])
ax2.set_xlabel("Time [$d$]",fontsize="large")
ax2.axhline(y=1.0,ls="dashed",c="teal",lw=1)
ax2.axhspan(ymin=-1.0E-7,ymax=1.0E-6,facecolor="pink",alpha=0.2)
ax2.set_yscale("log")
ax2.legend(title="From probe in the middle")
plt.show()
```
